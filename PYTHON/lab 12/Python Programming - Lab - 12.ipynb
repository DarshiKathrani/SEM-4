### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
mon=["Jan","Feb","Mar","Apr","May","Jun"]
visitors=[11000,12000,15000,12500,25000,17000]
bars=plt.bar(mon,visitors,0.5)
for i in bars:
    yc=i.get_height()
    plt.text(i.get_x(),yc,f"{yc}")
plt.show()
```
